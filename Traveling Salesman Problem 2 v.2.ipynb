```python
def nextPermutation(a):
    i, j = len(a)-1, len(a)-1
    
    while i > 0 and a[i] <= a[i-1]:
        i -= 1
        
    if i <= 0:
        return False
    
    i = i-1 #For matching index number
    
    while a[j] <= a[i]:
        j = j-1

    a[j], a[i] = a[i], a[j]
            
    j = len(a)-1 #the last index of the list
    i = i+1 #the index after the index switched before
            
    while j > i:
        a[i], a[j] = a[j], a[i]
        i += 1
        j -= 1
                
    return True    

def minCost (sumCost, L, M):
    temp = 0
    L2 = L[1:] + [L[0]]
    
    for i in range(len(L)):
        if M[L[i]][L2[i]] == 0:
            return False
        else:
            temp += M[L[i]][L2[i]]
        
    if sumCost[0] == 0 or sumCost[0] > temp:
        sumCost[0] = temp
    else:
        return False
    
    return True

n = int(input())
L = list(range(n))
sumCost = [0]

H = [list(map(int, input().split())) for _ in range(n)]

while True:
    minCost(sumCost, L, H)
    
    if not(nextPermutation(L)):
        break

print(sumCost[0])

```
